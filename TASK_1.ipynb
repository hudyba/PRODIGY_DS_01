```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df = pd.read_csv("iris.csv", names=column_names)
```

```python
print(df.head())
```

### BAR CHART FOR DISTRIBUTION OF CATOGORICAL AND CONTINUOUS VARIABLE:

```python
plt.figsize=(8, 5)
sns.countplot(x="species", data=df)
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Distribution of Iris Species")
plt.show()
```

### HISTOGRAM FOR DISTRIBUTION OF CATOGORICAL AND CONTINUOUS VARIABLE:

```python
plt.figsize=(10, 6)
sns.histplot(df["sepal_length"], bins=15, kde=True, color="skyblue")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Sepal Length Distribution in Iris Dataset")
plt.show()
```
